```python
import pandas as pd
import numpy as np
```

## Import and Merge Data

```python
movies = pd.read_csv('small data/movies.csv')
ratings = pd.read_csv('small data/ratings.csv')
```

```python
print("movies shape: {}".format(movies.shape))
print("first 5 movies\n")
movies.head(5)
```

```python

print("ratings shape: {}".format(ratings.shape))
print("first 5 ratings:\n")
ratings.head(5)
```

```python
# add movie title from movies to rating

final_data = pd.merge(movies,ratings, on="movieId", how="left")
final_data.head(5)
```

```python
final_data2 = pd.merge(ratings,movies, on="movieId", how="left")
final_data2.head(5)
```

```python
movie_data = final_data2
movie_data.head(5)
```

```python
# make genres into list format

movie_data['genres'] = movie_data['genres'].str.split('|')
movie_data.head(5)
```

```python
movie_data.shape
```

## Create 0.8 Train data and 0.2 Test data

```python
from sklearn.model_selection import train_test_split
```

```python
#split data into train and test
train_data, test_data = train_test_split(movie_data, test_size=0.2, stratify=movie_data.userId)
```

```python
train_data
```

```python
test_data
```

```python
# sort train data by user and movie
train_data = train_data.sort_values(['userId', 'movieId'])
train_data.head()
```

```python
test_data = test_data.sort_values(['userId','movieId'])
test_data.head()
```

## Save the train_data and test_Data

```python
train_data.to_csv('train_data.csv', index=False)
```

```python
test_data.to_csv('test_data.csv', index=False)
```
